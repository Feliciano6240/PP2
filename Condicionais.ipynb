## Estrutura match...case
O Python não possui estrutura *switch...case* como em outras linguagens, tal qual o JavaScript. Porém, a partir da versão 3.10, foi inserida na linguagem a estrutura *match...case*. Assim, os códigos do Problema 3 podem ser reescritos da seguinte forma:

```
plano = input("Digite plano (A, B ou C):")
salario = float(input("Digite o salario:"))
novo_salario = 0

match plano:
  case 'A':
    novo_salario = salario + salario*0.1
  case 'B':
    novo_salario = salario + salario*0.15
  case 'C':
      novo_salario = salario + salario*0.2

print("Seu novo salário é %f"%novo_salario)
```
